```python
# If 4 divides p-1, this algorithm returns a tuple with a probability of 1/2
# If 4 divides p-3, this algorithm return a tuple with a probability of 1/2 + 1/(p-1)
def test(p):
    if p not in Primes():
        raise ValueError()
        return
    big = p^2
    x = ZZ.random_element(1,p,distribution='uniform')
    y = ZZ.random_element(1,p,distribution='uniform')
    k = kronecker(-x^2-y^2,p)
    if (k == -1):
        raise ValueError()
        return 
    k = -x^2-y^2
    if (k%p == 0):
        if (k%big == 0):
            return (0,x+p,y,0)
        return (0,x,y,0)
    try:
        z = convert(int(mod(k,p).sqrt()),p)
    except TypeError:
        raise ValueError()
        return
    except ValueError:
        raise ValueError()
        return
    k = z^2-k
    if (k%big == 0):
        return (0,x+p,y,z)
    return (0,x,y,z)

def convert(n,p):
    if(n>=0):
        return n
    else:
        return p-n

# rounding to the nearest integer
def nearest(n):
    if n>=0:
        if n-int(n)>=0.5:
            return int(n)+1
        else:
            return int(n)
    else:
        if int(n)-n>=0.5:
            return int(n)-1
        else:
            return int(n)

# computes norm of an element
def distance(A,B):
    return ((A[0]-B[0])^2+(A[1]-B[1])^2+(A[2]-B[2])^2+(A[3]-B[3])^2)

# multiplies two quaternions
def MULL(A,B):
    P = [A[0]*B[0]-A[1]*B[1]-A[2]*B[2]-A[3]*B[3],A[0]*B[1]+A[1]*B[0]+A[2]*B[3]-A[3]*B[2],A[0]*B[2]+A[2]*B[0]+A[3]*B[1]-A[1]*B[3],A[0]*B[3]+A[3]*B[0]+A[1]*B[2]-A[2]*B[1]]
    return tuple(P)

# right euclidean algorithm in the Hurwitz order
def euclid(tup1,tup2,p):
    norm1 = distance(tup1,[0,0,0,0])
    norm2 = distance(tup2,[0,0,0,0])
    if (norm1 == 0 or norm2 == 0):
        return 0
    if norm1 == p:
        print(tup1)
        return 1
    if norm2 == p:
        print(tup2)
        return 1
    if norm1 >= norm2:
        big = tup1
        small = tup2
        m = norm2
    else:
        big = tup2
        small = tup1
        m = norm1
    P = [small[0]/m,-small[1]/m,-small[2]/m,-small[3]/m]
    g = MULL(P,big)
    prenew = [nearest(g[0]),nearest(g[1]),nearest(g[2]),nearest(g[3])]
    inter = MULL(small,prenew)
    newSmall = [big[0]-inter[0],big[1]-inter[1],big[2]-inter[2],big[3]-inter[3]]
    newSmall = tuple(newSmall)
    euclid(small,newSmall,p)

# Driver
def sumOfFour(p):
    try:
        tup = test(p)
    except ValueError:
        return
    euclid(tup,(p,0,0,0),p)
    return


sumOfFour(11587)
```
